
# What are the subclasses of CHEBI_35757 monocarboxylic acid anion with a *negative charge*, ordered by ascending *charge*

[link](https://terminology.nfdi4chem.de/ts/fuseki/dataset.html#query=PREFIX+rdfs%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2000%2F01%2Frdf-schema%23%3E%0APREFIX+owl%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2002%2F07%2Fowl%23%3E%0APREFIX+rdf%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F1999%2F02%2F22-rdf-syntax-ns%23%3E%0APREFIX+OBO%3A+%3Chttp%3A%2F%2Fpurl.obolibrary.org%2Fobo%2F%3E%0APREFIX+chebi%3A+%3Chttp%3A%2F%2Fpurl.obolibrary.org%2Fobo%2Fchebi%2F%3E%0APREFIX+xsd%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2001%2FXMLSchema%23%3E%0A%0ASELECT+%3Fs+%3Flabel+%3Fobject+%3Fcharge%0AWHERE+%7B%0A++%3Fs+rdfs%3AsubClassOf+OBO%3ACHEBI_35757%3B%0A+++++chebi%3Acharge+%3Fobject%3B%0A+++++rdfs%3Alabel+%3Flabel.++%0A++BIND(xsd%3Ainteger(%3Fobject)+as+%3Fcharge).%0A++FILTER+(+%3Fcharge+%3C+0)%0A%7D%0AORDER+BY+ASC(%3Fcharge)%0ALIMIT+50%0A%23+the+subclasses+of+CHEBI_35757+monocarboxylic+acid+anion+with+a+negative+charge%0A%23+ordered+)

```
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX OBO: <http://purl.obolibrary.org/obo/>
PREFIX chebi: <http://purl.obolibrary.org/obo/chebi/>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?s ?label ?object ?charge
WHERE {
  ?s rdfs:subClassOf OBO:CHEBI_35757;
     chebi:charge ?object;
     rdfs:label ?label.  
  BIND(xsd:integer(?object) as ?charge).
  FILTER ( ?charge < 0)
}
ORDER BY ASC(?charge)
LIMIT 50
```

## what are the ChEBI classes with the *lagest mass*?

[link](https://terminology.nfdi4chem.de/ts/fuseki/dataset.html#query=PREFIX+rdfs%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2000%2F01%2Frdf-schema%23%3E%0APREFIX+owl%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2002%2F07%2Fowl%23%3E%0APREFIX+rdf%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F1999%2F02%2F22-rdf-syntax-ns%23%3E%0APREFIX+OBO%3A+%3Chttp%3A%2F%2Fpurl.obolibrary.org%2Fobo%2F%3E%0APREFIX+chebi%3A+%3Chttp%3A%2F%2Fpurl.obolibrary.org%2Fobo%2Fchebi%2F%3E%0APREFIX+xsd%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2001%2FXMLSchema%23%3E%0A%0ASELECT+%3Fs+%3Flabel+%3Fmaxmass%0AWHERE%7B%0A++%7B%0A++%09SELECT++(MAX(%3Fmass)+as+%3Fmaxmass)%0A++%09WHERE+%7B%3Fs+chebi%3Amass+%3Fmass.%0A++++BIND(xsd%3Afloat(%3Fmass)+as+%3Fmass_f).%0A++++FILTER+(str(%3Fmass_f)+!%3D+'NaN').%0A++++%7D%0A++%7D%0A++++%7B%0A++++++++%3Fs+rdf%3Atype+owl%3AClass%3B%0A++++++++rdfs%3Alabel+%3Flabel%3B%0A++++++++chebi%3Amass+%3Fmaxmass.%0A++%7D%0A%7D%0A%0A%23+what+are+the+ChEBI+classes+with+the+lagest+mass%3F)


```SPARQL
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX OBO: <http://purl.obolibrary.org/obo/>
PREFIX chebi: <http://purl.obolibrary.org/obo/chebi/>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?s ?label ?maxmass
WHERE{
  {
  	SELECT  (MAX(?mass) as ?maxmass)
  	WHERE {?s chebi:mass ?mass.
    BIND(xsd:float(?mass) as ?mass_f).
    FILTER (str(?mass_f) != 'NaN').
    }
  }
    {
        ?s rdf:type owl:Class;
        rdfs:label ?label;
        chebi:mass ?maxmass.
  }
}
```
